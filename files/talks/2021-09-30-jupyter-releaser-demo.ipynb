## Why do we need it?

  - Complex release process for JupyterLab
  - More and more projects to maintain: https://pypi.org/manage/projects/
  - **Ensure best practices**:
    - Documented, transparent releases:
    - Assets are built on CI, not on a local machine
    - Process is streamlined, less error-proned
    - Full release logs are accessible: https://github.com/jtpio/jupyter_releaser/runs/3753501817?check_suite_focus=true
  - Motivation: https://jupyter-releaser.readthedocs.io/en/latest/background/theory.html#motivation

## How to adopt the Releaser?
  - Add the Releaser config: https://jupyter-releaser.readthedocs.io/en/latest/how_to_guides/convert_repo.html
  - Example PR: https://github.com/QuantStack/jupyterlab-snippets/pull/38
  - Add `npm` and `PyPI` tokens:
    - PyPI: https://pypi.org/manage/account/token/
    - npm: https://www.npmjs.com/settings/jtpio/tokens
    - GitHub Secrets: https://github.com/jtpio/jupyter_releaser/settings/secrets/actions
  - `PYPI_TOKEN_MAP`:
  
```bash
QuantStack/jupyterlab-snippets,pypi-AgEIcHlwaS5vcmcCJ...
jupyterlab/retrolab,pypi-AgEIcHlwaS5vcmcCJ...
...
```

## Workflow


- Example with https://github.com/QuantStack/jupyterlab-snippets/pull/39
- Draft Changelog
  - Go to: https://github.com/jtpio/jupyter_releaser/actions/workflows/draft-changelog.yml
  - Specify the new version spec
- Full Release
  - Go to: https://github.com/jtpio/jupyter_releaser/actions/workflows/full-release.yml
  - Use the parameters from the Changelog PR
